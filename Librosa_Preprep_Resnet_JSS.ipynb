Creating dataset for Genre recognition using CNN on melspectograms

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import pathlib
```

```python
cmap = plt.get_cmap('inferno')
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
datasets = "/content/gdrive/My Drive/datasets/Music/"
```

```python
#datasets+'genres/rock/rock.00035.wav'
# songname = datasets+"genres/metal/metal.00000.au"
songname = datasets+'genres/rock/rock.00035.wav'
```

```python
y , sr = librosa.load(songname, mono=True, duration=30)
spectogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, n_fft=2048, hop_length=1024)
spectogram = librosa.power_to_db(spectogram, ref=np.max)
plt.axes().set_aspect('equal')
plt.imshow(spectogram, origin="lower", cmap=cmap);
plt.axis('off');
```

```python
plt.figure(figsize=(10,10))
plt.specgram(y, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
plt.axis('off');
plt.savefig('demo.png')
```

```python
#songname = datasets+'genres/rock/rock.00035.wav'
plt.figure(figsize=(10,10))
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    pathlib.Path(datasets+f'img_data/{g}').mkdir(parents=True, exist_ok=True)     
    for filename in os.listdir(datasets+f'genres/{g}'):
        songname = datasets+f'genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=30)
        plt.specgram(y, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
        plt.axis('off');
        plt.savefig(datasets+f'img_data/{g}/{filename[:-3].replace(".", "")}.png')
        plt.clf()
    print(f'{g} done!')
```
